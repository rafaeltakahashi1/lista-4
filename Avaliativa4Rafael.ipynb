```python
# exercício 1
# os primeiros 16 dados do arquivo csv

import pandas as pd
arquivo = open('trabalho4/lista4.csv', 'r')
df = pd.read_csv(arquivo, sep=';')

df.head(16)
```

```python
# exercício 2
# desconto de 20% em todos os itens da classe "BMW"
lista = df['BMW']

def desconto(x):
    return x * 0.80

bmwdesc = list(map(desconto, lista))

print(bmwdesc)
```

```python
# exercicio 3
# código para analisar a lista(coluna BMW) se a informação passa a ser "muito caro" ou "razoavel" contando antes e depois do desconto feito anteriormente.
import csv

for x in lista:
    if x > 50000:
        print('muito caro')
    else:
        print('razoavel')
        
print()
print('Depois do desconto')
print()

for x in bmwdesc:
    if x > 50000:
        print('continua caro')
    else:
        print('razoavel')
```

```python
# exercício 4
# zipando 4 colunas do dataframe, contando com antes e depois do desconto da BMW e as colunas "mes" e "honda"
import csv

mes = df['mês']
honda = df['Honda']

list(zip(mes,honda,lista,bmwdesc))
```

```python
# exercicio 5 - Média e mediana
media = lista.mean()
print('média:',media)

mediana = lista.median()
print(mediana)
```
